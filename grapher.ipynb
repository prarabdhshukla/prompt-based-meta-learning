```python
import re
import matplotlib.pyplot as plt

def plot_accuracy_vs_steps(file_name):
    # Define the regex pattern with groups for step, loss, and accuracy
    pattern = r"step: \s+(\d+) \| loss: ([\d.]+), accuracy: ([\d.]+)%"

    # Read the contents of the file
    with open(file_name, 'r') as file:
        file_content = file.read()

    # Find all matches in the file content
    matches = re.findall(pattern, file_content)

    # Extracted values
    steps = []
    accuracies = []

    # Extract values for each group from the matched lines
    for match in matches:
        steps.append(int(match[0]))  # Convert step to integer
        accuracies.append(float(match[2]))  # Convert accuracy to float

    # Plotting the graph
    plt.figure(figsize=(8, 6))
    plt.plot(steps, accuracies, linestyle='-')
    plt.title('Accuracy vs Steps')
    plt.xlabel('Steps')
    plt.ylabel('Accuracy (%)')
    plt.grid(True)
    plt.tight_layout()

    # Show the plot
    plt.show()
```

```python
plot_accuracy_vs_steps('out.txt')
```
